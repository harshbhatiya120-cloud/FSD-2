```python
db.products.find({},{name:1,price:1,_id:0}).sort({price:-1})# price top to lower
db.products.find({},{name:1,price:1,_id:0}).sort({price:1})#price lower to top
db.products.find({},{name:1,price:1,_id:0}).sort({price:1}).limit(1)#lowest price
db.products.find({},{name:1,price:1,_id:0}).sort({price:-1}).limit(1)#higest price
```

```python
#funcation to find value
$eq,$lt,$lte,$gt,$gte,$ne,$in,$nin

db.products.find({price:{$eq:2500}})
db.products.find({price:{$gt:20000}})
db.products.find({price:{$gt:20000}},{name:1,_id:0})
db.products.find({qty:{$ne:3}})
db.products.find({price:{$nin:[750,2500]}})
db.products.find({price:{$in:[750,2700]}})


```

```python
#LOGICAL
$and,
$or,
$nor,
$not,

db.products.find({$and:[{cat:"electronics"},{price:{$gt:30000}}]})
db.products.find({price:{$not:{$lte:30000}})
```
